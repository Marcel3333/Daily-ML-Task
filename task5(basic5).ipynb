```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import itertools
from sklearn.metrics import adjusted_rand_score
from sklearn.mixture import GaussianMixture
```

```python
data = pd.read_csv('../Desktop/Ä‡wiczenia na KMeans/basic5.csv')
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.drop(columns = ['color']))
```

```python
plt.scatter(data['x'], data['y'], c = data['color'])
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(data_scaled)
```

```python
plt.scatter(data['x'], data['y'], c = kmeans.labels_)
plt.show()
```

```python
# ARI SCORE
ari = adjusted_rand_score(data['color'], kmeans.labels_)
print(f"Adjusted Rand Index: {ari}")
```
