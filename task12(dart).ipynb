```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import itertools
from sklearn.metrics import adjusted_rand_score
```

```python
data = pd.read_csv('../Desktop/ćwiczenia na KMeans/dart.csv')
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.drop(columns = ['color']))
plt.scatter(data['x'], data['y'], c = data['color'])
plt.show()
```

```python
dbscan = DBSCAN(eps = 0.2, min_samples = 1)
dbscan.fit(data_scaled)
ari = adjusted_rand_score(data['color'], dbscan.labels_)
print(f"Adjusted Rand Index: {ari}")
```
