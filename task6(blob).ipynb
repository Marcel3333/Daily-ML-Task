```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import itertools
from sklearn.metrics import adjusted_rand_score
from sklearn.mixture import GaussianMixture
```

```python
data = pd.read_csv('../Desktop/Ä‡wiczenia na KMeans/blob.csv')
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.drop(columns = ['color']))
```

```python
plt.scatter(data['x'], data['y'], c = data['color'])
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

gmm = GaussianMixture(n_components=4, covariance_type='full')
labels = gmm.fit_predict(data_scaled)

plt.scatter(data['x'], data['y'], c=labels, cmap='tab10', s=10)
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

```python
#Gaussian Mixture Score
ari = adjusted_rand_score(data['color'] , labels)
print(f"Adjusted Rand Index: {ari}")
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data_scaled)
plt.scatter(data['x'], data['y'], c = kmeans.labels_)
plt.show()
```

```python
#Kmeans score, very similar to the Gaussian
ari = adjusted_rand_score(data['color'], kmeans.labels_)
print(f"Adjusted Rand Index: {ari}")
```
